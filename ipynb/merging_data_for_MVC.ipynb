```python
comb_mat = np.add(np.multiply(feat_mat, .3), np.multiply(item_mat, .7))
```

```python
def con_cur_to_bgg_db():
    con = pg2.connect(host='34.216.22.202',
                  dbname='postgres',
                  user='postgres')
    cur = con.cursor(cursor_factory=RealDictCursor)
    return con, cur

def insert_sim(sim_mat):
    con, cur = con_cur_to_bgg_db()
    for i in range(10000):
        cur.execute("INSERT INTO rec_lookup (gs, sim) VALUES\
                ({}, '{{{}}}');".format(i, str(list(sim_mat[i]))[1:-1]))
    con.commit()
    con.close()
    
def insert_top_K(top_K_mat):
    con, cur = con_cur_to_bgg_db()
    for i in range(10000):
        cur.execute("UPDATE rec_lookup SET top_k = \
                '{{{}}}' WHERE gs={};".format(str(list(top_K_mat[i]))[1:-1], i))
    con.commit()
    con.close()
    
def insert_exp(exp_dict):
    con, cur = con_cur_to_bgg_db()
    for i in range(10000):
        if exp_dict[i]:
            cur.execute("UPDATE rec_lookup SET exp = \
                '{{{}}}' WHERE gs={};".format(str(list(exp_dict[i]))[1:-1], i))
    con.commit()
    con.close()
    
```
